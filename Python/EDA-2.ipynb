```python
def uniquevals(col):
    print(f'Details of the particular col {col} is : {df[col].unique()}')
    
def valuecounts(col):
    print(f'Valuecounts of the particular col {col} is : {df[col].value_counts()}')
```

```python
for i in df.columns:
    uniquevals(i)
    print()
    print(" - "*30)
    print()
```

```python
for i in df.columns:
    valuecounts(i)
    print()
    print("-"*40)
    print()
    
```
